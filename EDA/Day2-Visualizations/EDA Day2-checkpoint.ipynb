```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.linspace(0,10,20)
x
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,3))
axes[0].plot(x,x**2,x,np.exp(x))
axes[0].set_title('Normal')
axes[0].grid(True)

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Logarithmic Scale(y)')
axes[1].grid(True)
plt.show()
```

```python
fig,axes=plt.subplots()

axes.plot(x,x**2,color='b')
axes.set_ylabel('Area')
axes2=axes.twinx()
axes2.plot(x,x**3,color='r')
axes2.set_ylabel('Volume')
plt.show()
```

```python
n=np.array([0,1,2,3,4,5])
n
```

```python
fig,axes=plt.subplots(1,4,figsize=(11,3))
axes[0].scatter(x,x+np.random.randn(len(x)))
axes[0].set_title('Scatter')

axes[1].step(n,n**2)
axes[1].set_title('Step')

axes[2].bar(n,n**2)
axes[2].set_title('Bar')

axes[3].fill_between(x,x**2,x**3,color='g')
axes[3].set_title('fill_between')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
mm=MinMaxScaler()
data_mm=mm.fit_transform((df1['Purchase']).values.reshape(-1,1))
df1['Purchase_mm']=data_mm
df1[['Purchase','Purchase_mm']].head()
```

```python
ss=StandardScaler()
data_ss=ss.fit_transform((df1['Purchase']).values.reshape(-1,1))
df1['Purchase_ss']=data_ss
df1[['Purchase','Purchase_mm','Purchase_ss']].head()
```

```python
df1[['Purchase','Purchase_mm','Purchase_ss']].describe()
```

```python
fig,ax=plt.subplots(1,3,figsize=(10,3))
sns.distplot(df1['Purchase'],ax=ax[0])
sns.distplot(df1['Purchase_mm'],ax=ax[1])
sns.distplot(df1['Purchase_ss'],ax=ax[2])
plt.show()
```
