### 1.7 Imputing the missing values from the given dataset?

```python
# we will impute the categorical variables with 0 or 1 and continuous variables with median value



c.info()
```

```python
# Imputing the missing values from First sexual intercourse


print(c['First sexual intercourse'].describe())
c['First sexual intercourse'].fillna(str(c['First sexual intercourse'].median()), inplace=True)
```

```python
# Imputing the missing values from Num of pregnancies


print(c['Num of pregnancies'].describe())
c['Num of pregnancies'].fillna(str(c['Num of pregnancies'].median()), inplace=True)
```

```python
# Imputing the missing values from Smokes


print(c['Smokes'].describe())
c['Smokes'].fillna('0.0', inplace=True)
```

```python
# Imputing the missing values from Smokes (years)


print(c['Smokes (years)'].describe())
c['Smokes (years)'].fillna(str(c['Smokes (years)'].median()), inplace=True)
```
