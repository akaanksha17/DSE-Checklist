```python
sns.violinplot(y='MonthlyIncome',x='SeriousDlqin2yrs',data=df1)
```

```python
q1=df1['MonthlyIncome'].quantile(0.25)
q2=df1['MonthlyIncome'].quantile(0.5)
q3=df1['MonthlyIncome'].quantile(0.75)

def inc_cat(mi):
    if mi<q1:
        return 1
    elif q1<=mi<q2:
        return 2
    elif q2<=mi < q3:
        return 3
    else:
        return 4
df1['mi_cat']=df1['MonthlyIncome'].apply(inc_cat)
```

```python
df1.groupby('mi_cat')['MonthlyIncome'].agg(['min','max'])
```

```python
pd.crosstab(df1['mi_cat'],df1['SeriousDlqin2yrs'],normalize=0)
```

```python
ct=pd.crosstab(df1['mi_cat'],df1['SeriousDlqin2yrs'],normalize=0)
ct.plot.bar(stacked=True)
plt.show()
```

The chart shows that default rate of lowest income category is 8.6% while default rate of highest income category is 4.8%
