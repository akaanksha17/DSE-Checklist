```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=[3,5,2,8,9,2,2,2,7]
```

```python
import statistics as st
```

```python
mean=np.mean(data)
mean
```

```python
median=np.median(data)
median
```

```python
mode=st.mode(data)
mode
```

```python
plt.figure(figsize=(10,5))
plt.hist(data)
plt.axvline(mean,color='g')
plt.axvline(median,color='b')
plt.axvline(mode,color='r')
plt.show()
```

```python
mean=df['GRE Score'].mean()
median=df['GRE Score'].median()
mode=df['GRE Score'].mode()[0]
plt.figure(figsize=(10,5))
plt.hist(df['GRE Score'],bins=100)
plt.axvline(mean,color='y')
plt.axvline(median,color='b')
plt.axvline(mode,color='r')
plt.xlabel('GRE Score')
plt.ylabel('Freq')
plt.show()
```
